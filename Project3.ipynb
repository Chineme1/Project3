```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as si
import scipy.optimize as so
```

(a)

```python
a, b = 0.05, 1.2
h = 0.1
t = np.arange(0, 100 + h, h)
Y0 = [1, 1]

def Y(X, t): # ATP = X[0], F6P = x[1]
    A = -X[0] + a*X[1] + X[1]*X[0]**2
    F = b - a*X[1] - X[1]*X[0]**2
    return [A, F]

Y = si.odeint(Y, Y0, t)

plt.plot(t, Y[:, 0], 'r--', label = 'ATP')
plt.plot(t, Y[:, 1], 'bo', label = 'F6P')
plt.legend()
plt.show()
```

```python
a, b1, b2 = 0.05, 1.2, 0.8 # b1 from (0, 30), b2 from (30, 100)
h = 0.1
t1 = np.arange(0, 30 + h, h)
t2 = np.arange(30, 100 + h, h)
Y0 = [1, 1]

def Y1(X, t1): # ATP = X[0], F6P = x[1]
    A = -X[0] + a*X[1] + X[1]*X[0]**2
    F = b1 - a*X[1] - X[1]*X[0]**2
    return [A, F]

def Y2(X, t2): # ATP = X[0], F6P = x[1]
    A = -X[0] + a*X[1] + X[1]*X[0]**2
    F = b2 - a*X[1] - X[1]*X[0]**2
    return [A, F]

Y1 = si.odeint(Y1, Y0, t1)
Y2 = si.odeint(Y2, Y1[-1], t2)

plt.plot(t1, Y1[:, 0], 'r--', label = 'ATP')
plt.plot(t2, Y2[:, 0], 'r--')
plt.plot(t1, Y1[:, 1], 'bo', label = 'F6P')
plt.plot(t2, Y2[:, 1], 'bo')
plt.legend()
plt.show()
```

```python
a = 0.05
h = 0.1
t = np.arange(0, 200 + h, h)
Y0 = [1, 1, 1]

def Y(X, t): # b = X[0], ATP = x[1], F6P = X[2]
    b = -0.1*X[2]*X[1]*np.exp(-1/X[0])
    A = -X[1] + a*X[2] + X[2]*X[1]**2
    F = X[0] - a*X[2] - X[2]*X[1]**2
    return [b, A, F]

Y = si.odeint(Y, Y0, t)

plt.plot(t, Y[:, 1], 'r--', label = 'ATP')
plt.plot(t, Y[:, 2], 'bo', label = 'F6P')
plt.legend()
plt.show()
```
