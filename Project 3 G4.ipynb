Glycolysis is the biochemical process by which living cells break down sugars to release
energy, which is used to form Adenosine Triphosphate (ATP). The complete glycolytic pathway
is a complex series of ten enzyme-catalyzed reactions, involving intermediates that are
derivatives of common sugars like fructose and glucose. Here, we wish to examine glycolytic
oscillations that arise is one of these intermediate processes that relate ATP and Fructose-6-
phosphate (F6P). 

By measuring the concentrations of ATP and F6P in yeast and muscle extracts, scientists have
demonstrated that their rates of change depend nonlinearly on the current concentration of the
other. A simplified model for this part of glycolysis is: 

d[ATP]
dt = − [ATP] + a[F6P] + [ATP]**2
[F6P]

d[F6P]
dt = b − a[F6P] − [ATP]**2
[F6P]

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
a = 0.05
b = 1.2


def APt(x,y,t): return -1*x + a*y + y*x**2
def FPt(x,y,t): return b -1*a*y - y*x**2
h = 0.5
t = np.arange(0,100+h,h)
x = np.zeros(t.shape)
y = np.zeros(t.shape)
x[0] = 1
y[0] = 1
#I'maa make a Runge-Kunta method 4 cause it's more accurate.
# I did and modified the code from the tutorial and got this.
for i in range(len(t)-1):
    kx1 = APt(x[i], y[i], t[i])
    ky1 = FPt(x[i], y[i], t[i])
    
    kx2 = APt(x[i]+ h*kx1/2, y[i] + h*ky1/2,t[i]+h/2)
    ky2 = FPt(x[i]+ h*kx1/2, y[i] + h*ky1/2,t[i]+h/2)
    
    kx3 = APt(x[i]+ h*kx2/2, y[i] + h*ky2/2,t[i]+h/2)
    ky3 = FPt(x[i]+ h*kx2/2, y[i] + h*ky2/2,t[i]+h/2)
    
    kx4 = APt(x[i]+ h*kx3, y[i] + h*ky3,t[i]+h/2)
    ky4 = FPt(x[i]+ h*kx3, y[i] + h*ky3,t[i]+h/2)
    
    kx = (kx1 +2*kx2 + 2*kx3 +kx4)/6
    ky = (ky1 + 2*ky2 + 2*ky3 + ky4)/6
    
    x[i+1] = x[i] + h*kx
    y[i+1] = y[i] + h*ky

plt.figure(figsize=(15,10))   
plt.plot(t,x,'b-', label='APt/dt')
plt.plot(t,y,'r', label='FPt/dt')
plt.xlabel('Time(seconds)',size ='30')
plt.ylabel('Height',size ='30')
plt.legend()
plt.grid()
plt.show()

#AP=
#FP =
```
